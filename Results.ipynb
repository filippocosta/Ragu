```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import subprocess
import numpy as np
from Bio import SeqIO
```

```python
#Results from k*d10
accuracy = [0.66,0.76,0.90,0.96,0.98,1]
accuracy_nok = [0.56, 0.59, 0.64, 0.69, 0.71, 0.76]
kmer = [5,6,7,8,9,10]
plt.figure(figsize = (10,7))
plt.plot(kmer,accuracy, '--', color = 'crimson', marker = 'o', markersize = 10, label = 'With Kmer')
plt.plot(kmer,accuracy_nok, '--', color = 'steelblue' ,marker = 'o', markersize = 10, label = 'Without Kmer')
plt.title('ACCURACY - K*D10', fontsize = 20, weight = 'bold')
plt.xlabel('Word length', fontsize = 18, labelpad = 15, weight = 'semibold')
plt.ylabel('Test accuracy', fontsize = 18, labelpad = 15, weight = 'semibold')
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.grid()
plt.legend(fontsize = 'x-large')
plt.show()
```

```python
#Results from k10d*
accuracy = [0.50,0.96,0.98,0.99,1,1]
emb = [1,2,3,4,10,11]
plt.figure(figsize = (10,7))
plt.plot(emb,accuracy, '--', color = 'crimson', marker = 'o', markersize = 10)
plt.title('ACCURACY - K10D*', fontsize = 20, weight = 'bold')
plt.xlabel('Embedding dimension', fontsize = 18, labelpad = 15, weight = 'semibold')
plt.ylabel('Test accuracy', fontsize = 18, labelpad = 15, weight = 'semibold')
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.grid()
plt.show()
```
